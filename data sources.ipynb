```python
# Dependencies
import json
import pandas as pd
from pprint import pprint
import ijson
from sqlalchemy import create_engine
```

```python
# url = "https://apps.who.int/gho/athena/data/GHO/AIR_5,AIR_41,AIR_42.json?profile=simple&filter=COUNTRY:*;REGION:*"
# filename = "file://localhost/Users/Endur/Desktop/13-ETL-Project/data/air.json" 
filename = "data/air.json" 
```

```python
with open(filename, 'r') as f:
    objects = ijson.items(f, 'fact')
    columns = list(objects)

dim_list=[]
val_list=[]

c=1
for n in columns[0]:
#     if c1==0:
        pprint(n["dim"])
        dim_list.append(n["dim"])
        val_list.append(n["Value"])
      
        c+=1
```

```python
who_df = pd.DataFrame(dim_list)
who_df.head(400)
```

```python
who_df.insert(loc=0, column='Deaths', value=val_list)
```
